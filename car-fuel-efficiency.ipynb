# standard scalar

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
```

```python
X_train_std= sc.fit_transform( X_train)
```

```python
pickle.dump(sc , open( "./scalar.pkl", "wb"))
```

# dl model!!!!!!!!

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense
```

```python
X_train.keys()
```

```python
#initialising ANN
model = Sequential()


#add input layer
model.add(Dense(units= 64, activation = "relu",input_shape = [len(X_train.keys() ) ]))

#add 2nd hidden layer layer
model.add(Dense(units= 64, activation = "relu"))

#add 3rd hidden layer
model.add(Dense(units= 64, activation = "relu"))

#adding the output layer
model.add(Dense(units= 1))
```

```python
#compile the ann
model.compile(loss= 'mean_squared_error', optimizer= 'Adam', metrics= ['mean_squared_error','mean_absolute_error'])
```

```python
model.summary()
```

# training the model

```python
model.fit(X_train_std, Y_train, batch_size= 100, epochs= 1000)
```

```python
model.evaluate(X_train_std, Y_train)
```
